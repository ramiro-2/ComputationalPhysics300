```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Number of rolls
N0=1000000
# a c m selection
a=953
c=21
m=1000000
# seed selection for each dice
s1=9
s2=6
D1=[]
D2=[]
# Normilization E=f(a,c,m), P=f(dice), S=f(m)
# E tweaks distribution boundry 
E=1
# P determines values outputted by rolls
P=1/6
# S scales number generation to desired P range
S=10**6*P*E
# Indexing incidents of double rolls
N1=0
for i in range(N0):
    # Rolling first Dice
    s1 = ((a*s1+c)%m)
    X1 = np.ceil(s1/S)
    D1.append(X1)
    # Rolling second Dice
    s2 = ((a*s2+c)%m)
    X2 = np.ceil(s2/S)
    D2.append(X2)
    # Indexing doubles
    if X1==X2 and X1==6:
        N1=N1+1

plt.plot(D1,"b.",markersize=0.5)
plt.show()
plt.hist(D1)
plt.show()
#print(D1)
plt.plot(D2,"b.",markersize=0.5)
plt.show()
plt.hist(D2)
plt.show()
#print(D2)
print('Double chance is:',N1/N0*100,'%')
```
