```python
import numpy as np
import matplotlib.pyplot as plt
import os
data = np.loadtxt("circular.txt",float)
plt.imshow(data)
plt.savefig('data.jpg')
print('This is the original image:')
plt.show()
data="data.jpg"
data=os.stat(data)
print(f'The file size is {data.st_size} bytes.')
```

```python
data = np.loadtxt("circular.txt",float)
DataC=np.fft.fft(data)
dataC=np.absolute(DataC)
plt.imshow(dataC)
plt.savefig('dataC.jpg')
print('Here we see a representation of the image in compression via Fourier transform:')
plt.show()
dataC="dataC.jpg"
dataC=os.stat(dataC)
print(f'While the image is completely lost, the file size is smaller')
print(f'only {dataC.st_size} bytes.')
```

```python
DataR=np.fft.ifft(DataC).real
plt.imshow(DataR)
plt.savefig('dataR.jpg')
print('If we reverse our Fourier transform we get the following:')
plt.show()
dataR="dataR.jpg"
dataR=os.stat(dataR)
print(f'The file remains at {dataR.st_size} Bytes.')
print('This highlights the convenience of Fourier transforming data')
print('that needs to be transmitted.')
```
