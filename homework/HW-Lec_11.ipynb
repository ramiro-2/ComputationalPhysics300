```python
import matplotlib.pyplot as plt
import numpy as np
from random import random

fig1 = plt.figure()
ax1 = fig1.add_subplot(111, aspect='equal')

x = np.linspace(0,2,5001)

def y1(x):
    return np.sqrt(1-(x-1)**2)
ax1.plot(x,y1(x),'k')

def y2(x):
    return 2-np.sqrt(4-x**2)
ax1.plot(x,y2(x),'k')



ax1.set_xlim(0,2)
ax1.set_ylim(0,2)
ax1.fill_between(x, y1(x), y2(x), where=y1(x)>y2(x), facecolor='green')
plt.grid(1,which='both')
```

```python
def y3(x):
    return y1(x)-y2(x)
R=np.where(y3(x) > 0)
R=R[0]
R=R[-1]
xMax=x[R+1]
print('xMax=',xMax)
yMax=np.max(y3(x))
print('yMax=',yMax)
plt.plot(x,y3(x))
plt.xlim(0,xMax)
plt.ylim(0,yMax)
plt.grid(1,which='both')
```

```python
for N in [10,100,1000,10000,100000,1000000]:
    C=0
    gx=[]
    gy=[]
    rx=[]
    ry=[]
    for i in range(N):
        x = xMax*random()
        #print('x=',x)
        y = yMax*random()
        #print('y=',y)
        if y<y3(x):
            C=C+1
            #print(C)
            gx.append(x)
            gy.append(y)
        else:
            #print('nope')
            rx.append(x)
            ry.append(y)
    x = np.linspace(0,2,5001)
    plt.plot(x,y3(x),'k',gx,gy,'g.',rx,ry,'r.',markersize=0.5)
    plt.xlim(0,xMax)
    plt.ylim(0,yMax)
    plt.grid(1,which='both')
    I=xMax*yMax*C/N
    plt.show()
    p=I/(xMax*yMax)
    var=N*p*(1-p)
    std=np.sqrt(var)/N
    print('I=',I)
    print('Variance=',var)
    print('Standard Deviation=',std)
    
```
