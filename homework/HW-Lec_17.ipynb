```python
import numpy as np

pi=np.pi
inf=np.tan(pi/2)
eps,sig=1,1

def LJ(r,eps,sig):
    E=4*eps*((sig/r)**12-(sig/r)**6)
    return E

def ri(x0,xf,y0,yf):
    r=np.sqrt((xf-x0)**2+(yf-y0)**2)
    return r

def GridMin(x0,xf,y0,yf,N=10):
    X=np.linspace(x0,xf,N)
    Y=np.linspace(y0,yf,N)
    a=np.array([X,Y,X,Y,X,Y])
    E=0
    for i in range(N):
        for j in range(N):
            for k in range(N):
                # r12
                r=ri(a[0,i],a[2,j],a[1,i],a[3,j])
                if r==0:
                    E12=inf
                else:
                    E12=LJ(r,eps,sig)
                # r13
                r=ri(a[0,i],a[4,k],a[1,i],a[5,k])
                if r==0:
                    E13=inf
                else:
                    E13=LJ(r,eps,sig)
                # r12
                r=ri(a[2,j],a[4,k],a[3,j],a[5,k])
                if r==0:
                    E23=inf
                else:
                    E23=LJ(r,eps,sig)
                Eijk=E12+E13+E23
                if Eijk<E:
                    E=Eijk
                    x1=a[0,i]
                    y1=a[1,i]
                    x2=a[2,j]
                    y2=a[3,j]
                    x3=a[4,k]
                    y3=a[5,k]
    return E, x1, y1, x2, y2, x3, y3

def RandMin(x0,xf,y0,yf,N=100,dx=0.1,dy=0.1):
    M=GridMin(x0,xf,y0,yf,10)
    E=M[0]
    for i in range(N):
        x1=M[1]+2*dx*np.random.random()-dx
        y1=M[2]+2*dy*np.random.random()-dy
        x2=M[3]+2*dx*np.random.random()-dx
        y2=M[4]+2*dy*np.random.random()-dy
        x3=M[5]+2*dy*np.random.random()-dy
        y3=M[6]+2*dy*np.random.random()-dy
        
        r12=ri(x1,x2,y1,y2)
        r13=ri(x1,x3,y1,y3)
        r23=ri(x2,x3,y2,y3)
        
        E12=LJ(r12,eps,sig)
        E13=LJ(r13,eps,sig)
        E23=LJ(r23,eps,sig)
        Eijk=E12+E13+E23
        if Eijk<E:
            E=Eijk
    return E, x1, y1, x2, y2, x3, y3
```

```python
G=GridMin(-1,1,-1,1)
print('Grid Minimum Report:')
print('Min of',G[0],'at:')
print('x1=',G[1])
print('y1=',G[2])
print('x2=',G[3])
print('y2=',G[4])
print('x3=',G[5])
print('y3=',G[6])
R=RandMin(-1,1,-1,1,100000,1,1)
print(' ')
print('Bounded Random Minimum report:')
print('Min of',R[0],'at:')
print('x1=',R[1])
print('y1=',R[2])
print('x2=',R[3])
print('y2=',R[4])
print('x3=',R[5])
print('y3=',R[6])
#print(G[0],'vs',R[0])
```
